```python
#Importing packages for the analysis
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os
%matplotlib inline
```

```python
#To check working directory
os.getcwd()
```

```python
#To import covid-19 data for Africa
#Data was collected at 08:02 GMT 11-01-2021 from bbc.co.uk
coa= pd.read_csv('COVID_AFRICA.csv')
```

```python
coa.head()
```

```python
coa.info()
```

```python
coa.describe()
```

```python
#To change the column heads
coa.columns=['Country', 'TotalCases','ActiveCases', 'Recovered', 'TotalDeaths']
```

```python
coa.head()
```

```python
#To change the columns to integers
coa.Country=coa.Country.astype("category")
coa.TotalCases=coa.TotalCases.astype("int")
coa.ActiveCases=coa.ActiveCases.astype('int')
coa.Recovered=coa.Recovered.astype('int')
coa.TotalDeaths=coa.TotalDeaths.astype('int')
```

```python
coa.info()
```

TOP 10 AFRICAN COUNTRIES WITH COVID-19 CASES

```python
#To get the top 10 countries wih the highest number of cases
#Create a dataframe that has been sorted by country and totalcases
top10_confirmed=coa[['Country','TotalCases']].sort_values('TotalCases',ascending=False).nlargest(10,'TotalCases')
```

```python
top10_confirmed
```

```python
top10_confirmed.info()
```

```python
top10_confirmed=coa[['Country','TotalCases']].sort_values('TotalCases',ascending=False).nlargest(10,'TotalCases')
```

```python
top10_confirmed
```

```python
fig,ax=plt.subplots()
fig.set_size_inches(15,15)
sns.set_style('darkgrid')
V1=sns.barplot(data=top10_confirmed, x=top10_confirmed.index,y='TotalCases')
plt.ylabel('TotalCases', fontsize=20)
plt.xlabel('Countries', fontsize=20)
plt.title('TOP 10 AFRICAN COUNTRIES WITH COVID-19 CASES',fontsize=30,color='Green',fontname='DejaVu Sans')
plt.show()

```
